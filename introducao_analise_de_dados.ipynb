# Os números alarmantes de mortes 

Vemos que um número grande de mulheres morrem nas clínicas, de 1841 a 1846

Vamos analisar a proporção de mortes usando uma função onde faremos a primeira
alteração no dado. criaremos uma coluna chamada proporcao_de_mortes que vai
receber a razão entre mortes e nascimentos, deaths/births.

```r
dado_anual <- dado_anual %>% 
  mutate(proporcao_de_mortes = deaths/births)
```

##Ah o PIPE! %>%

PIPE = TUBO

E é isso o que ele faz, ele representa um tubo que leva o dado da esquerda para a direita. Ou seja, 

`dado_anual` *(nossa tabela)* `%>%` *(pipe - transfere o conteúdo da tabela para a função seguinte)* `mutate(nome_da_coluna_nova_criada = recebe modificação_no_dado)` *(função do tidyverse que cria colunas modificadas, alteradas, mutadas)*

```r
# Imprimir o objeto dado_anual para verificar a nova coluna
dado_anual
```

Note que temos agora uma coluna com a proporção de mortes por nascimento

Agora que você preparou o dado podemos plotar proporcao_de_mortes pelo tempo. No plot abaixo vamos observar como o efeito da lavagem de mãos alterou o número de mortes na clínica.(fazer group = 1)

# O efeito da lavagem da mãos observado!

A partir de **1847**, lavar as mãos se tornou obrigatório, e é possível observar isso no gráfico e ainda destacar esse período.

A partir desta **Data**, lavar as mãos se tornou obrigatório

```r
lavar_maos_comeca <- as.Date('1847-06-01')
```

Vamos adicionar uma coluna de VERDADEIRO ou FALSO (TRUE/FALSE) chamada lavar_maos_comeca

```r
mensal <-  mensal %>%
  mutate(lavar_maos_comeca = date >= lavar_maos_comeca)
```

Agora vamos plotar a proporção mensal de mortes antes e depois de começar a  lavagem de mãos obrigatória 


```r
ggplot(monthly, aes(x = date, y = , color = lavar_maos_comeca)) +
  geom_line() +
  labs(x = "Data", y = "Prop. de mortes")
```

# Mais lavagem de mãos = a menos mortes?

O gráfico mostrou que lavar as mãos teve um grande efeito no número de mortes, mas quanto em média de proporção de mortes foi reduzido graças a lavar as mãos?

Calculando a média da proporção de mortes antes e depois de lavar as mãos:

```r
sumario_mensal <- mensal %>%
  group_by(lavar_maos_comeca) %>%
  summarise(media_da_proporcao_de_mortes = mean(proporcao_de_mortes))
```

```r
# Imprimir o sumário
sumario_mensal
```

Wow! Parece que depois da obrigatoriedade de lavar as mãos o número de mortes
foi reduzido em 0.8! O que significa que antes morriam 10% das mulheres, e
depois da obrigatoriedade de lavar as mãos passaram a morrer 2%! (o que ainda é um número alto para os padrões atuais).
